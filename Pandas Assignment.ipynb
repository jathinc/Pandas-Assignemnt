```python
############################## Pandas Assignment #########################################################

import numpy as np
import pandas as pd

# Define the dataframe

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
                   'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                   'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
df
```

```python
# Fill in these missing numbers in FlightNumber column
df['FlightNumber'] = [10045.0, 10055.0, 10065.0, 10075.0, 10085.0]
df
```

```python
# make the FlightNumber column an integer column
df['FlightNumber'] = df['FlightNumber'].apply(lambda x : int(x))
df
```

```python
# split the From_To column by '_' and create separate From and To columns

df['From'] = df['From_To'].apply(lambda x : x.split('_')[0])
df['From']
```

```python
df['To'] = df['From_To'].apply(lambda x : x.split('_')[1])
df['To']
```

```python
# Standardize the From and To columns so that only the first letter is uppercase (e.g. "londON" should become "London".)

df['From'] = df['From'].apply(lambda x : x.capitalize())
df['From']
```

```python
df['To'] = df['To'].apply(lambda x : x.capitalize())
df['To']
```

```python
# Delete the From_To column from the data frame. 

del df['From_To']
df
```

```python
# Split the RecentDelays column to separate delay columns. 

#df['RecentDelay1'] = df['RecentDelays'].apply(lambda x : np.nan if x[0] is None else x[0])

#df['RecentDelay1'] = df['RecentDelays'].apply(lambda x : x.str[0])
df['delay_1'] = df['RecentDelays'].str[0]

df['delay_2'] = df['RecentDelays'].str[1]

df['delay_3'] = df['RecentDelays'].str[2]

df
```

```python
# Delete the RecentDelays column from the data frame. 

del df['RecentDelays']
df
```
